```python
#leman model for my reference 
import pandas as pd
df = pd.read_csv('ab_testing.csv')
df['Conversion_Numeric'] = df['Conversion'].apply(lambda x: 1 if x == 'Yes' else 0)
#print("Original DataFrame (df) head with Conversion_Numeric:")
#print(df.head())
#print("\nOriginal DataFrame (df) shape:", df.shape)
df_group_A = df[df['Group'] == 'A'].copy()
df_group_B = df[df['Group'] == 'B'].copy()
print("\n--- df_group_A (Group A) ---")
print("Head:")
print(df_group_A.head())
print("Shape:", df_group_A.shape)

print("\n--- df_group_B (Group B) ---")
print("Head:")
print(df_group_B.head())
print("Shape:", df_group_B.shape)
conversions_A = df_group_A['Conversion_Numeric'].sum()
conversions_B = df_group_B['Conversion_Numeric'].sum()

print("\n--- Conversion Counts ---")
print(f"Number of conversions in Group A: {int(conversions_A)}")
print(f"Number of conversions in Group B: {int(conversions_B)}")
total_users_A = len(df_group_A)
total_users_B = len(df_group_B)

print("\n--- Total Users per Group ---")
print(f"Total users in Group A: {total_users_A}")
print(f"Total users in Group B: {total_users_B}")
conversion_rate_A = (conversions_A / total_users_A) * 100
conversion_rate_B = (conversions_B / total_users_B) * 100
print(f"Group A Conversion Rate: {conversion_rate_A:.2f}%")
print(f"Group B Conversion Rate: {conversion_rate_B:.2f}%")
conversion_rate_difference = conversion_rate_B - conversion_rate_A
print(f"Conversion Rate Difference (Group B - Group A): {conversion_rate_difference:.2f} percentage points")

if conversion_rate_A > conversion_rate_B:
    print("dont go with the new model stick to the old model")
else:
    print("go with the new model ")
```
